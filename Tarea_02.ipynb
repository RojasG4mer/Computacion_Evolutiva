# Primer problema
\Un comerciante acude a cierto mercado a comprar naranjas con 500 Pesos. Le ofrecen dos tipos de naranjas: las de tipo A a 0,5 Pesos el kg y las de tipo B a 0,8 Pesos el kg. Sabemos que solo dispone en su camioneta de espacio para transportar 700 kg de naranjas como máximo y que piensa vender el kilo de naranjas de tipo A a 0,58 Pesos y el de tipo B a 0,9 Pesos. ¿Cuántos kilogramos de naranjas de cada tipo deberá comprar para obtener beneficio máximo?

Variables de desición:
- $x_A$ = kilos de naranjas A
- $x_B$ _ kilos de naranjas B

Ganancia por kilogramo para cada tipo de naranja:
$$ Gan_A = \text{ Precio venta A - Precio compra A } = 0.58 - 0.5 = 0.08 $$
$$ Gan_B = \text{ Precio venta B - Precio compra B } = 0.9 - 0.8 = 0.1 $$
Ganancia total de ambas naranajas:
$$ G = 0.08 x_A + 0.1 x_B $$

Condiciones:
- Dinero máximo 500:
$$ 0.5 x_A + 0.8 x_b \leq 500 $$
- Peso máximo 700 kg:
$$ x_A + x_B = 700 $$
- Compras positivas, pero no negativas o cero:
$$ x_A, x_B \geq 0$$

```python
import numpy as np

def optimizar_naranjas():
    print("--- Solución al Problema de las Naranjas ---")
    
    # Restricciones de forma Ax + By = C
    # 1. 0.5*x_A + 0.8*x_B = 500 (Presupuesto)
    # 2. 1.0*x_A + 1.0*x_B = 700 (Capacidad)
    
    # Lista para almacenar todos los vértices posibles (intersecciones)
    vertices_potenciales = []
    
    # 1. Origen
    vertices_potenciales.append(np.array([0.0, 0.0]))
    
    # 2. Intersecciones con los ejes (x_A = 0 o x_B = 0)
    # Para la restricción de presupuesto:
    vertices_potenciales.append(np.array([500.0 / 0.5, 0.0]))
    vertices_potenciales.append(np.array([0.0, 500.0 / 0.8]))
    
    # Para la restricción de capacidad:
    vertices_potenciales.append(np.array([700.0 / 1.0, 0.0]))
    vertices_potenciales.append(np.array([0.0, 700.0 / 1.0]))
    
    # 3. Intersección entre las dos rectas (Resolución del sistema 2x2 desde cero)
    a1, b1, c1 = 0.5, 0.8, 500.0
    a2, b2, c2 = 1.0, 1.0, 700.0
    
    determinante = (a1 * b2) - (a2 * b1)
    if determinante != 0:
        x_A_int = (c1 * b2 - c2 * b1) / determinante
        x_B_int = (a1 * c2 - a2 * c1) / determinante
        vertices_potenciales.append(np.array([x_A_int, x_B_int]))

    # Evaluar los vértices factibles
    mejor_beneficio = -1.0
    mejor_compra = None
    
    # Tolerancia para evitar errores de precisión de punto flotante
    tol = 1e-6 

    for vertice in vertices_potenciales:
        x_A = vertice[0]
        x_B = vertice[1]
        
        # Validar si el vértice cumple todas las restricciones (factibilidad)
        cond_negatividad = (x_A >= -tol) and (x_B >= -tol)
        cond_presupuesto = (0.5 * x_A + 0.8 * x_B) <= (500.0 + tol)
        cond_capacidad = (x_A + x_B) <= (700.0 + tol)
        
        if cond_negatividad and cond_presupuesto and cond_capacidad:
            # Calcular función objetivo
            beneficio = 0.08 * x_A + 0.10 * x_B
            
            if beneficio > mejor_beneficio:
                mejor_beneficio = beneficio
                mejor_compra = (x_A, x_B)

    print(f"La compra óptima es: {round(mejor_compra[0], 2)} kg de Tipo A y {round(mejor_compra[1], 2)} kg de Tipo B.")
    print(f"El beneficio máximo obtenido será de: {round(mejor_beneficio, 2)} Pesos.\n")

optimizar_naranjas()
```

# Segundo problema
Disponemos de monedas de distintos valores: 1, 2 pesos y monedas de 5, 10, 20 y 50 centavos. Supondremos una cantidad ilimitada de cada una.

Construir un algoritmo que dada una cantidad $P$ devuelva esa cantidad con monedas de esos tipos, usando un número mínimo de monedas.

```python
def algoritmo_goloso_monedas(cantidad_pesos):
    print(f"--- Solución Algoritmo Goloso para {cantidad_pesos} Pesos ---")
    
    # Convertimos los valores disponibles a centavos para operar con precisión
    # 2 pesos (200), 1 peso (100), 50, 20, 10 y 5 centavos.
    # El orden debe ser descendente para garantizar tomar el candidato más "prometedor" primero.
    denominaciones_centavos = np.array([200, 100, 50, 20, 10, 5])
    
    # Convertimos la cantidad solicitada a centavos
    cantidad_restante = int(round(cantidad_pesos * 100))
    
    # Partimos de una solución vacía
    solucion = []
    
    # Iniciamos el ciclo mientras la cantidad no sea cero
    while cantidad_restante > 0:
        moneda_seleccionada = None
        
        # Función de Selección: Buscar la moneda más grande que no exceda el restante
        for moneda in denominaciones_centavos:
            # Factibilidad: ¿Es válido tomar esta moneda?
            if moneda <= cantidad_restante:
                moneda_seleccionada = moneda
                break # Rompemos porque encontramos la mayor posible (el arreglo está ordenado)
        
        if moneda_seleccionada is None:
            print("No se puede encontrar una solución exacta con las denominaciones actuales.")
            break
            
        # Añadir a la solución y restar a la cantidad pendiente
        solucion.append(moneda_seleccionada / 100.0) # Guardamos el formato en pesos
        cantidad_restante -= moneda_seleccionada
        
    # Agrupar e imprimir resultados
    valores_unicos, conteos = np.unique(solucion, return_counts=True)
    
    print("Desglose óptimo de monedas a entregar:")
    for valor, cantidad in zip(reversed(valores_unicos), reversed(conteos)):
        if valor >= 1.0:
            print(f"- {cantidad} moneda(s) de {int(valor)} Peso(s)")
        else:
            print(f"- {cantidad} moneda(s) de {int(valor * 100)} centavos")

# Prueba con la cantidad propuesta en el documento (ejemplo 3.89, redondeado a múltiplos de 5 centavos si es necesario, pero usaremos 3.85 para exactitud con las monedas dadas, o puedes colocar el monto que desees evaluar)
algoritmo_goloso_monedas(4.85)
```
