# Búsqueda tabú

Es un procedimiento metaheurístico cuya característica distintiva es el uso de memoria adaptativa y de estrategias especiales de resolución de problemas.

Su filosofía se bas aen la explotación de diversas estrategias inteligentes para la resolución de problemas, basadas en procedimientos de aprendizaje.

El marco de memoria adaptativa de TS explota la historia del proceso de resolución del problema haciendo referencia a 4 dimensiones principales, consistentes en la
- propiedad de ser reciente.
- en frecuencia
- en calidad

Es una metaheurística que guía un proceso de búsqueda local con el propósito de explorar el espacio de soluciones más allá de subespacioes con óptimos locales, aplicando mecanismos que evitan regresar a subespacioes ya viistados.

Es una combinación de exploración sensible y memoria adaptativa.

- No utiliza población, en cada ciclo opera sobre una solución. Es un algoritmo de trayectoria.
- Trabaja con estrategias intensificadoras.

# Algoritmos genéticos

Combinan estructuras de información que codifican soluciones del problema que mejor se adapte a la función objetivo, costo o aptitud que se desea optimizar

Cada iteración se genera una población de individuos en base a los que mejor se adaptaron en la generación anterior.

No hacen busqueda aleatoria, explotan eficientemente la información histórica para generar nuevos puntos de búsqueda

1. Los AG trabajan con un conjutno de parámetrs codificados, no con los paraámetros en si mismos.
2. Los algoritmos genéticos buscan la solución a partir de una población de puntos y no desde un punto simple.
3. Los algoritmos genéticos utilizan la información de la función objetivo y no de derivada
4. Los AG utilizan la regla de tensión probabilísitca y no determinística.

## Terminología
Un cromosoma es dividido conceptualmente en genes, los cuales son bloques funcionales de DNA, cada uno de los cuales codifican una caracterísitica particular de cada individuo.
